### ðŸ›¸ Sentiment Analysis

```python
df_fake['titleText'] = df_fake['title'] + ' ' + df_fake['text']
df_true['titleText'] = df_true['title'] + ' ' + df_true['text']

df_fake['sent'] = 0
df_true['sent'] = 1

df = pd.concat([df_fake, df_true])
#karÄ±ÅŸtÄ±rmak iÃ§in
df = df.sample(frac=1).reset_index(drop=True)

df = df.drop(['title', 'text', 'subject', 'date'], axis = 1)

df.head(10)
```

```python
def clean_data(df):
    df['titleText']=df['titleText'].str.lower()
    df['titleText']=df['titleText'].str.replace('[^\w\s]','')
    df['titleText']=df['titleText'].str.replace('\d+','')
    df['titleText']=df['titleText'].str.replace('\n','').replace('\r','')
    
    return df
```

```python
clean_data(df)
```

```python
tdf=df[["titleText","sent"]].sample(100)
```

```python
tdf
```

```python
def detect_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
tdf['sentiment']=tdf['titleText'].apply(detect_sentiment)
```

```python
tdf
```

```python
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer(r"\w+")
tdf['titleText']=tdf['titleText'].apply(regexp.tokenize)
tdf.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

tdf['titleText'] = tdf['titleText'].apply(lambda x: [item for item in x if item not in stopwords.words('english')])
```

```python
tdf['titleText'] = tdf['titleText'].apply(lambda x: ' '.join([w for w in x if len(w)>2]))
```

```python
pos = tdf[tdf['sent'] == 1]
neg = tdf[tdf['sent'] == 0]
```

```python
pos_sentiment_words = ''.join([word for word in pos['titleText']]) 
neg_sentiment_words = ''.join([word for word in neg['titleText']])
```

```python
list_pos_words = [ x for x in pos_sentiment_words.split()]
freq_dis_pos = nltk.FreqDist(list_pos_words)
freq_dataframe = pd.DataFrame({'Words': list(freq_dis_pos.keys()), 'Count': list(freq_dis_pos.values())})
freq_dataframe = freq_dataframe.nlargest(columns="Count", n = 10)
```

```python
plt.figure(figsize=(16,5))
ax = sns.barplot(data=freq_dataframe, x= "Words", y = "Count")
ax.set(ylabel = 'Count')
ax.set(xlabel = 'Top 10 words used in true context')
plt.title("Top 10 words in the news labeled as True")
plt.show()
```

```python
list_neg_words = [ x for x in neg_sentiment_words.split()]
freq_dis_pos = nltk.FreqDist(list_neg_words)
freq_dataframe = pd.DataFrame({'Words': list(freq_dis_pos.keys()), 'Count': list(freq_dis_pos.values())}) 
freq_dataframe = freq_dataframe.nlargest(columns="Count", n = 10)
```

```python
plt.figure(figsize=(16,5))
ax = sns.barplot(data=freq_dataframe, x= "Words", y = "Count")
ax.set(ylabel = 'Count')
ax.set(xlabel = 'Top 10 words used in fake context')
plt.title("Top 10 words in the news labeled as Fake")
plt.show()
```

```python
def wc(data,bgcolor,title):
    plt.figure(figsize=(20,20))
    wc=WordCloud(background_color=bgcolor,stopwords=STOPWORDS)
    wc.generate(data)
    plt.imshow(wc,interpolation="bilinear")
    plt.title(title)
    plt.axis("off")
```
